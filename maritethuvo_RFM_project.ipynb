```python
import pandas as pd
```

```python
df = pd.read_excel('ecommerce retail.xlsx')
df
```

```python
df.info()
```

```python
df['CustomerID']= df['CustomerID'].dropna()
df['CustomerID']= df['CustomerID'].astype('Int64')
df['InvoiceDate']= pd.to_datetime(df['InvoiceDate']).dt.date
df
```

```python
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]
df = df[df['Quantity'] > 0]
df = df[df['UnitPrice'] > 0]
df
```

```python
df['TotalPrice']= df['Quantity'] * df['UnitPrice']
df_rfm= df.groupby('CustomerID').agg({'InvoiceNo': 'nunique', 'TotalPrice': 'sum','InvoiceDate':'max'}).reset_index()
df_rfm.columns = ['CustomerID', 'Frequency', 'Monetary','MaxInvoiceDate']
df_rfm
```

```python
df_rfm['MaxInvoiceDate'] = pd.to_datetime(df_rfm['MaxInvoiceDate'], errors='coerce')
snapshot_date= pd.to_datetime('2011-12-31')
df_rfm['Recency'] = (snapshot_date - df_rfm['MaxInvoiceDate']).dt.days
del df_rfm['MaxInvoiceDate']
df_rfm
```

```python
df_rfm.describe()
```

```python
df_rfm['F_score']= pd.qcut(df_rfm['Frequency'].rank(method='first'), q=5, labels=[1,2,3,4,5]).astype(int)
df_rfm['M_score']= pd.qcut(df_rfm['Monetary'].rank(method='first'), q=5, labels=[1,2,3,4,5]).astype(int)
df_rfm['R_score']= pd.qcut(df_rfm['Recency'].rank(method='first'), q=5, labels=[5,4,3,2,1]).astype(int)
df_rfm
```

```python
df_rfm['RFM_score']= df_rfm['R_score'].astype(str) + df_rfm['F_score'].astype(str) + df_rfm['M_score'].astype(str)
df_rfm
```

```python
segmentation_dict = {
    "Champions": ["555","554","544","545","454","455","445"],
    "Loyal": ["543","444","435","355","354","345","344","335"],
    "Potential Loyalist": [
        "553","551","552","541","542","533","532","531",
        "452","451","442","441","431","453","433","432",
        "423","353","352","351","342","341","333","323"],
    "New Customers": ["512","511","422","421","412","411","311"],
    "Promising": [
        "525","524","523","522","521","515","514","513",
        "425","424","413","414","415","315","314","313"],
    "Need Attention": ["535","534","443","434","343","334","325","324"],
    "About To Sleep": ["331","321","312","221","213","231","241","251"],
    "At Risk": [
        "255","254","245","244","253","252","243","242",
        "235","234","225","224","153","152","145","143",
        "142","135","134","133","125","124"],
    "Cannot Lose Them": ["155","154","144","214","215","115","114","113"],
    "Hibernating customers": ["332","322","233","232","223","222","132","123","122","212","211"],
    "Lost customers": ["111","112","121","131","141","151"]}
```

```python
segment_lookup = {segment: name
    for name, segments in segmentation_dict.items()
    for segment in segments}
```

```python
df_rfm['Segment'] = df_rfm['RFM_score'].map(segment_lookup)
```

```python
df_rfm
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
segment_counts = df_rfm['Segment'].value_counts().sort_values(ascending=False)
plt.bar(segment_counts.index, segment_counts.values)
plt.xticks(rotation=45, ha='right')
plt.title("Customer Count by RFM Segment")
plt.xlabel("Segment")
plt.ylabel("Number of Customers")
plt.tight_layout()
plt.show()
```

```python
r_heat = pd.crosstab(df_rfm['Segment'], df_rfm['R_score'])
plt.figure(figsize=(12,6))
sns.heatmap(r_heat, annot=True, fmt='d', cmap='Blues')
plt.title("Segment vs R_Score")
plt.ylabel("Segment")
plt.xlabel("R_Score")
plt.show()
```

```python
f_heat = pd.crosstab(df_rfm['Segment'], df_rfm['F_score'])
plt.figure(figsize=(12,6))
sns.heatmap(f_heat, annot=True, fmt='d', cmap='Greens')
plt.title("Segment vs F_Score")
plt.ylabel("Segment")
plt.xlabel("F_Score")
plt.show()
```

```python
m_heat = pd.crosstab(df_rfm['Segment'], df_rfm['M_score'])
plt.figure(figsize=(12,6))
sns.heatmap(m_heat, annot=True, fmt='d', cmap='Oranges')
plt.title("Segment vs M_Score")
plt.ylabel("Segment")
plt.xlabel("M_Score")
plt.show()
```

# Phân tích hiện trạng

1. Champions = R=4–5, F=4–5, M=4–5
Những KH này tập trung ở mức điểm cao
-> profit engine

2. Hibernating & Lost Customers tập trung ở R=1–2

Recency thấp = họ đã lâu không quay lại.

Lost customers: tập trung cực mạnh tại R=1 (490), M thấp, F thấp

Hibernating: R=2–3 nhưng k có ở nhóm điểm M cao.

-> đây là nhóm ngủ quên, không mua hàng, có nguy cơ chết hẳn.

3. At Risk nằm ở R=1–2 nhưng có F và M khá cao

Họ từng mua nhiều.

Họ tiêu tiền không ít.

Nhưng gần đây họ không quay lại.

-> mất họ là mất tiền thật

4. Potential Loyalist nằm rải ở R=3–4 nhưng M trung bình

-> là nhóm sẵn sàng convert thành Loyal hoặc Champions.

5. New Customers (R=4–5 nhưng F & M thấp)

-> store đã thu hút được khách, nhưng chưa phát triển để tăng LTV

# CTA

1. Tập trung vào Loyal và Champions vì những người này đã sẵn sàng chi tiền cho mình

* Cross-sell, upsell

* Bundle discount

* Early access sản phẩm mới


2. Lấy lại nhóm khách đang At risk
* Winback campaigns (email, SMS)

* Personal discount based on last category purchased

* “We miss you” offers

* Extended warranty / free shipping to push a comeback purchase

3. Kéo dẫn nhóm Potential Loyalist
* Promote membership program

* Lower-bar offers ("Buy again, get 10%")
